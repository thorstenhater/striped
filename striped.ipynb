```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as wdg
```

```python
fn = "/Users/hater/Downloads/1280px-Commodore_Grace_M._Hopper,_USN_(covered).jpg"
img = plt.imread(fn)
```

```python
fg, ax = plt.subplots()
ax.imshow(img)
ax.axis('off')
```

```python
def split(arr, width=1, axis=0):
    arrs = np.array_split(arr, arr.shape[axis]//(width + 1), axis=axis)
    return np.concat(arrs[0::2], axis=axis), np.concatenate(arrs[1::2], axis=axis)
```

```python
fg, axs = plt.subplots(2, 2, figsize=(10, 10))
def show(width):
    print(width)
    l, r = split(img, width=width, axis=0)
    ul, dl = split(l, width=width, axis=1)
    ur, dr = split(r, width=width, axis=1)

    axs[0, 0].imshow(ul)
    axs[0, 0].axis('off')
    axs[0, 1].imshow(ur)
    axs[0, 1].axis('off')
    axs[1, 0].imshow(dl)
    axs[1, 0].axis('off')
    axs[1, 1].imshow(dr)
    axs[1, 1].axis('off')
    plt.subplots_adjust(wspace=0.01, hspace=0.01)

wdg.interact(show, width=wdg.IntSlider(5, 5, 50, 5))
```
